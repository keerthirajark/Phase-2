```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
file_path = r"C:\Users\Welcome\OneDrive\Pictures\Desktop\transit_data_500_rows.csv"
df = pd.read_csv(file_path)
df['date'] = pd.to_datetime(df['date'], dayfirst=True)
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='vehicle_type', palette='pastel')
plt.title('Vehicle Type Distribution')
plt.xlabel('Vehicle Type')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
plt.figure(figsize=(6,4))
sns.barplot(data=df, x='time_of_day', y='fare_collected', estimator='mean', palette='coolwarm')
plt.title('Avg Fare Collected by Time of Day')
plt.ylabel('Average Fare Collected')
plt.tight_layout()
plt.show()
plt.figure(figsize=(6,4))
sns.barplot(data=df, x='weather_condition', y='delay_minutes', estimator='mean', palette='Blues')
plt.title('Avg Delay by Weather Condition')
plt.ylabel('Average Delay (minutes)')
plt.tight_layout()
plt.show()
df['total_passengers'] = df['boarding_count'] + df['alighting_count']
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='day_type', y='total_passengers', palette='Set2')
plt.title('Total Passengers by Day Type')
plt.ylabel('Total Passengers')
plt.tight_layout()
plt.show()
top_routes = df.groupby('route_name')['fare_collected'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(8,4))
top_routes.plot(kind='bar', color='orange')
plt.title('Top 10 Routes by Total Fare Collected')
plt.ylabel('Total Fare Collected')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
daily_delay = df.groupby('date')['delay_minutes'].mean()
plt.figure(figsize=(10,4))
daily_delay.plot(marker='o', color='crimson')
plt.title('Average Daily Delay Over Time')
plt.ylabel('Avg Delay (minutes)')
plt.xlabel('Date')
plt.grid(True)
plt.tight_layout()
plt.show()
```
